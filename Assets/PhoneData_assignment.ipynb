```python
dl = pd.read_xml('AccelerationData.xml')
```

```python
print(dl)
```

```python
time = dl["time"]
x = dl["x"]
y = dl["y"]
z = dl["z"]

# Vi laver plottet
plt.figure(figsize=(10, 6))

# Plot x vs time
plt.plot(time, x, label="X")

# Plot y aksen vs time på samme plot
plt.plot(time, y, label="Y")

# Plot z aksen vs time samme plot
plt.plot(time, z, label="Z")

# Sætter labels og titler
plt.xlabel("Time")
plt.ylabel("Values")
plt.title("Plot of X, Y, and Z vs Time")

# Adder legend (den der box oppe til højre)
plt.legend()

# Viser plottet med nogle conditions plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
